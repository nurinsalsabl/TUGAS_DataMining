```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import KNNImputer

scaler = MinMaxScaler()
knn_cols1 = ['scoma','charges','hday', 'dementia', 'alb', 'bili',
             'slos', 'totmcst', 'dnrday','wblc', 'totcst', 'crea',
             'glucose',  'bun', 'adlp', 'sps', 'diabetes','ph',
             'prg6m', 'prg2m', 'surv6m', 'surv2m']
df[knn_cols1] = scaler.fit_transform(df[knn_cols1])

knn_imputer = KNNImputer(n_neighbors=5)
df[knn_cols1] = knn_imputer.fit_transform(df[knn_cols1])
df[knn_cols1] = scaler.inverse_transform(df[knn_cols1])

mean_cols = [
    'aps', 'adlsc', 'pafi', 'num.co', 'avtisst', 'resp', 'sod', 'temp',
    'meanbp', 'hrt', 'edu'
]
df[mean_cols] = df[mean_cols].apply(lambda x: x.fillna(x.mean()))

# knn_cols = []
# knn_imputer = KNNImputer(n_neighbors=5)
# df[knn_cols] = knn_imputer.fit_transform(df[knn_cols])

print("\nMissing values setelah imputasi:")
print(df.isnull().sum())
```

```python
from scipy.stats import chi2_contingency

cat_cols = df.select_dtypes(include=['object']).columns
chi2_results = {}

for col in cat_cols:
    contingency_table = pd.crosstab(df[col], df['death'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    chi2_results[col] = p

chi2_results = pd.DataFrame(chi2_results.items(), columns=['Feature', 'P-Value'])
chi2_results = chi2_results.sort_values(by='P-Value')

print(chi2_results)
```

```python
for col in cat_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, hue='death', palette='coolwarm')
    plt.title(f"Distribusi Death berdasarkan {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
import numpy as np

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

cramers_results = {col: cramers_v(df[col], df['death']) for col in cat_cols}
cramers_results = pd.DataFrame(cramers_results.items(), columns=['Feature', 'CramerV'])
cramers_results = cramers_results.sort_values(by='CramerV', ascending=False)

print(cramers_results)
```

```python
import pandas as pd

df_original = pd.read_csv("data.csv")
df_cleaned = pd.read_csv("cleaned_data.csv")

missing_features = list(set(df_original.columns) - set(df_cleaned.columns))

print(f"Fitur yang hilang setelah cleaning: {missing_features}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

df = pd.read_csv("cleaned_data.csv")

y = df["death"]
X = df.drop(columns=["death", "id"])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

k = 5
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi KNN (k={k}): {accuracy:.4f}")
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative (0)', 'Positive (1)'], yticklabels=['Negative (0)', 'Positive (1)'])

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Dapatkan probabilitas prediksi positif
y_prob = knn.predict_proba(X_test)[:, 1]

# Hitung ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Garis diagonal sebagai baseline
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

ROC Curve (kurva biru) Menunjukkan trade-off antara True Positive Rate (TPR) dan False Positive Rate (FPR).
Diagonal abu-abu Garis acuan untuk model random guessing (AUC = 0.5).

AUC = 0.99  Artinya model hampir sempurna dalam memprediksi kelas (sAalahnya dikit)

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Hitung probabilitas prediksi positif
y_prob = knn.predict_proba(X_test)[:, 1]

# Hitung Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_prob)
average_precision = average_precision_score(y_test, y_prob)

# Plot Precision-Recall Curve
plt.figure(figsize=(6, 6))
plt.plot(recall, precision, color='blue', lw=2, label=f'AP Score = {average_precision:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.grid()
plt.show()
```

Precision tetap tinggi (~1.0) di sebagian besar Recall
Model hampir tidak salah dalam memprediksi positif.
Saat Recall mendekati 1.0, Precision mulai turun Artinya, jika model mencoba menangkap semua kasus positif, ia mulai membuat lebih banyak kesalahan (False Positives meningkat).
AP Score = 0.99 Model sangat baik dalam menyeimbangkan
